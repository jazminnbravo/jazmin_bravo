```python
players = ["Valentina", "Mateo", "Camila", "Santiago", "Lucía"]

stats = {}
for p in players:
    stats[p] = {"total": 0, "wins": 0, "best": 0, "rounds": 0}
```

```python
round_number = 1

for r in rounds:
    print(f"Ronda {round_number} - {r['theme']}:")

    scores_round = {}

    for player, judges in r["scores"].items():
        total = Calculos.calcular_total(judges)

        scores_round[player] = total

        Actualizar_Jugador.actualizar_(stats, player, total)

    winner = Ranking.obtener_ganador(scores_round)
    stats[winner]["wins"] += 1

    print(f"Ganador: {winner} ({scores_round[winner]} pts)")

    ordenados = Ranking.ordenar_tabla(stats)
    Imprimir.imprimir_tabla(ordenados)

    round_number += 1
```
